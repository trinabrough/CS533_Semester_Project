## Using BERT for Inference

```python
def computeSimilarities(df):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    l1 = BertModel.from_pretrained("bert-base-uncased")
    for index, row in df.iterrows():
        q1 = row['question_1']
        token = tokenizer.encode_plus(q1)
        input_ids = torch.tensor(token['input_ids']).unsqueeze(0)
        attention_mask = torch.tensor(token['attention_mask']).unsqueeze(0)
        output_1 = l1(input_ids=input_ids, attention_mask=attention_mask)
        hidden_state = output_1[0]
        pooler = hidden_state[:, 0]

        q2 = row['question_2']
        token = tokenizer.encode_plus(q2)
        input_ids = torch.tensor(token['input_ids']).unsqueeze(0)
        attention_mask = torch.tensor(token['attention_mask']).unsqueeze(0)
        output_2 = l1(input_ids=input_ids, attention_mask=attention_mask)
        hidden_state = output_2[0]
        pooler2 = hidden_state[:, 0]

        cos_sim = F.cosine_similarity(pooler, pooler2, dim=1)
        df.at[index, 'cosine_sim'] = cos_sim.item()
        #df['cosine_sim'].iloc[index] = cos_sim.item()
```

```python
computeSimilarities(df)
```

```python
df.head(5)
```

```python
def classify(df):
    results = pd.DataFrame(columns=['Threshold', 'Accuracy', 'Precision', 'Recall'])
    i = 0
    for threshold in range(70, 99, 1):  # Changed range values to integers
        threshold /= 100  # Convert threshold to float
        df['pred'] = 0
        df.loc[df['cosine_sim'] >= threshold, 'pred'] = 1
        accuracy = accuracy_score(df['label'], df['pred'])
        precision = precision_score(df['label'], df['pred'])
        recall = recall_score(df['label'], df['pred'])
        results.loc[i] = [threshold, accuracy, precision, recall]
        i += 1
        
    results = results.set_index('Threshold')
    return results


results = classify(df)
print(results)
results.to_csv('classification_results.csv')
```

### Torch Dataset
- Tokenizes questions
- Prepares to interact with DataLoader

```python
class MultiLabelDataset(torch.utils.data.Dataset):

    def __init__(self, text1, text2, labels, tokenizer, max_len):
        self.tokenizer = tokenizer
        self.text1 = text1
        self.text2 = text2
        self.targets = labels
        self.max_len = max_len

        # Determine the length of the dataset based on the larger of the two text inputs
        self.length = max(len(self.text1), len(self.text2))

    def __len__(self):
        return self.length

    def __getitem__(self, index):
        text1 = self.text1[index]
        text2 = self.text2[index]

        inputs1 = self.tokenizer.encode_plus(
            text1,
            None,
            add_special_tokens=True,
            max_length=self.max_len,
             padding='max_length',
            truncation=True,
            return_token_type_ids=True
        )
        inputs2 = self.tokenizer.encode_plus(
            text2,
            None,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_token_type_ids=True
        )

        input_ids1 = inputs1['input_ids']
        attention_mask1 = inputs1['attention_mask']
        token_type_ids1 = inputs1["token_type_ids"]

        input_ids2 = inputs2['input_ids']
        attention_mask2 = inputs2['attention_mask']
        token_type_ids2 = inputs2["token_type_ids"]

        return {
            'input_ids1': torch.tensor(input_ids1, dtype=torch.long),
            'attention_mask1': torch.tensor(attention_mask1, dtype=torch.long),
            'token_type_ids1': torch.tensor(token_type_ids1, dtype=torch.long),
            'input_ids2': torch.tensor(input_ids2, dtype=torch.long),
            'attention_mask2': torch.tensor(attention_mask2, dtype=torch.long),
            'token_type_ids2': torch.tensor(token_type_ids2, dtype=torch.long),
            'targets': torch.tensor(self.targets[index], dtype=torch.float)
        }
```

### Bert Class
- First layer is BERT pretrained
- Adding additional layers
- Using Siamese Neural Network

```python
class SiameseBERT(torch.nn.Module):
    def __init__(self, num_classes):
        super(SiameseBERT, self).__init__()
        self.bert = BertModel.from_pretrained("bert-base-uncased")
        self.classifier = torch.nn.Linear(768 * 2, num_classes)  # Concatenating the representations of two inputs
        self.dropout = torch.nn.Dropout(0.1)
        self.softmax = torch.nn.Softmax(dim=1)

    def forward(self, input_ids1, attention_mask1, token_type_ids1, input_ids2, attention_mask2, token_type_ids2):
        # Encoding the first input
        output1 = self.bert(input_ids=input_ids1, attention_mask=attention_mask1, token_type_ids=token_type_ids1)
        pooled_output1 = output1.pooler_output
        # Encoding the second input
        output2 = self.bert(input_ids=input_ids2, attention_mask=attention_mask2, token_type_ids=token_type_ids2)
        pooled_output2 = output2.pooler_output

        # Concatenate the pooled outputs
        concatenated = torch.cat((pooled_output1, pooled_output2), dim=1)
        concatenated = self.dropout(concatenated)

        # Classification layer
        logits = self.classifier(concatenated)

        return self.softmax(logits)
```

### Train Function

```python
def loss_fn(outputs, targets):
    return torch.nn.BCELoss()(outputs, targets)

def train(model, training_loader, optimizer, loss_fn, device):
    model.train()  # Set the model to training mode
    total_loss = 0.0  # Initialize the total loss
    
    # Iterate over the training data loader
    for data in tqdm(training_loader):
        # Extract the input data and target labels from the current batch
        input_ids1 = data['input_ids1'].to(device)
        attention_mask1 = data['attention_mask1'].to(device)
        token_type_ids1 = data['token_type_ids1'].to(device)
        input_ids2 = data['input_ids2'].to(device)
        attention_mask2 = data['attention_mask2'].to(device)
        token_type_ids2 = data['token_type_ids2'].to(device)
        targets = data['targets'].to(device)

        # Forward pass: compute predicted outputs by passing the input data through the model
        outputs = model(input_ids1, attention_mask1, token_type_ids1,
                        input_ids2, attention_mask2, token_type_ids2)

        # Calculate the loss
        loss = loss_fn(outputs, targets)

        # Zero the gradients
        optimizer.zero_grad()

        # Backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()

        # Perform a single optimization step (parameter update)
        optimizer.step()

        # Update the total loss
        total_loss += loss.item()

    # Calculate the average loss over all batches
    average_loss = total_loss / len(training_loader)

    return average_loss
```

### Evaluate Function

```python
def validation(model, testing_loader, device):
    model.eval()  # Set the model to evaluation mode
    fin_targets = []
    fin_outputs = []
    
    with torch.no_grad():
        for data in tqdm(testing_loader):
            # Extract the input data and target labels from the current batch
            targets = data['targets']
            input_ids1 = data['input_ids1'].to(device)
            attention_mask1 = data['attention_mask1'].to(device)
            token_type_ids1 = data['token_type_ids1'].to(device)
            input_ids2 = data['input_ids2'].to(device)
            attention_mask2 = data['attention_mask2'].to(device)
            token_type_ids2 = data['token_type_ids2'].to(device)

            # Forward pass: compute predicted outputs by passing the input data through the model
            outputs = model(input_ids1, attention_mask1, token_type_ids1,
                            input_ids2, attention_mask2, token_type_ids2)
            
            # Apply sigmoid activation to the outputs and move them to CPU
            outputs = torch.sigmoid(outputs).cpu().detach()
            
            # Extend the lists with the outputs and targets
            fin_outputs.extend(outputs)
            fin_targets.extend(targets)
    
    # Stack the lists to form tensors
    fin_outputs = torch.stack(fin_outputs)
    fin_targets = torch.stack(fin_targets)
    
    return fin_outputs, fin_targets
```

### Training setup

- hyperparameters
- setup dataset
- setup parameters
- setup dataloader

```python
MAX_LEN = 128
BATCH_SIZE = 64
EPOCHS = 3
NUM_OUT = 2 # binary task
LEARNING_RATE = 2e-05
# loss_fn = ContrastiveLoss()
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

training_data = MultiLabelDataset(X_train['question_1'].to_numpy(), X_train['question_2'].to_numpy(), labels_train, tokenizer, MAX_LEN)
test_data = MultiLabelDataset(X_test['question_1'].to_numpy(), X_test['question_2'].to_numpy(), labels_test, tokenizer, MAX_LEN)

train_params = {'batch_size': BATCH_SIZE,
                'shuffle': True,
                'num_workers': 0
                }

test_params = {'batch_size': BATCH_SIZE,
                'shuffle': False,
                'num_workers': 0
                }    

training_loader = torch.utils.data.DataLoader(training_data, **train_params)
testing_loader = torch.utils.data.DataLoader(test_data, **test_params)
```

### Train,  Evaluate

- model.to -> send to GPU, if available (anything computed should be put onto the GPU)
- setup optimizer - ADAM 
- for each epoch, train, show the loss, evaluate on the test data

```python
model = SiameseBERT(NUM_OUT)
model.to(device)    

optimizer = torch.optim.Adam(params =  model.parameters(), lr=LEARNING_RATE)

for epoch in range(EPOCHS):
    loss = train(model, training_loader, optimizer, loss_fn, device) 
    print(f'Epoch: {epoch}, Loss:  {loss}')  
    guess, targs = validation(model, testing_loader, device)
    guesses = torch.max(guess, dim=1)
    targets = torch.max(targs, dim=1)
    print('accuracy on test set {}'.format(accuracy_score(guesses.indices, targets.indices)))
    print('recall on test set {}'.format(recall_score(guesses.indices, targets.indices)))
    print('precision on test set {}'.format(precision_score(guesses.indices, targets.indices)))
```

## Classification with Pre-trained Open Source Model: 

```python


# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased-finetuned-mrpc")
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased-finetuned-mrpc")

# Define classes
classes = ["not paraphrase", "is paraphrase"]

# Define sequences (group of pairs)
sequences = [(row['question_1'], row['question_2']) for _, row in df.iterrows()]

# Set batch size
batch_size = 4  

# Process sequences in batches
softmax_probs_all = []
for i in range(0, len(sequences), batch_size):
    batch_sequences = sequences[i:i+batch_size]
    tokenized_sequences = tokenizer(batch_sequences, return_tensors="pt", padding=True, truncation=True)
    classification_logits = model(**tokenized_sequences)[0]
    softmax_probs = torch.softmax(classification_logits, dim=1).tolist()
    softmax_probs_all.extend(softmax_probs)
```

```python
threshold = .5
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Initialize lists to store predicted labels and actual labels
predicted_labels = []
actual_labels = []

# Iterate over each sequence and its corresponding softmax probabilities
for i, softmax_probs in enumerate(softmax_probs_all):
    # Get the actual label from the DataFrame
    actual_label = df['label'].iloc[i]
    
    # Determine predicted class based on the softmax probabilities and threshold
    predicted_label = 1 if softmax_probs[1] >= threshold else 0
    
    # Append predicted and actual labels to the lists
    predicted_labels.append(predicted_label)
    actual_labels.append(actual_label)

# Compute accuracy, precision, and recall using scikit-learn functions
accuracy = accuracy_score(actual_labels, predicted_labels)
precision = precision_score(actual_labels, predicted_labels)
recall = recall_score(actual_labels, predicted_labels)

# Print results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
